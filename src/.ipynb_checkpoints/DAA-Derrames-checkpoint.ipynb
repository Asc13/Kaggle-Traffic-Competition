# Tratamento de dados

***
- Convers√£o de dados 
***

```python
print(df['Residence_type'].unique())

def convert_Residence_type(residence_type):
    if residence_type == 'Urban':
        return 0
    elif residence_type == 'Rural':
        return 1
    
df['Residence_type'] = df['Residence_type'].apply(convert_Residence_type)  
print(df['Residence_type'].unique())
```

```python
print(df['smoking_status'].unique())

def convert_smoking_status(status):
    if status == 'Unknown':
        return 0
    elif status == 'never smoked':
        return 1
    elif status == 'smokes':
        return 2
    elif status == 'formerly smoked':
        return 3
    
df['smoking_status'] = df['smoking_status'].apply(convert_smoking_status)  
print(df['smoking_status'].unique())
```

```python
print(df['work_type'].unique())

def convert_work_type(status):
    if status == 'Private':
        return 0
    elif status == 'Self-employed':
        return 1
    elif status == 'Govt_job':
        return 2
    elif status == 'children':
        return 3
    elif status == 'Never_worked':
        return 4
df['work_type'] = df['work_type'].apply(convert_work_type)  
print(df['work_type'].unique())
```

```python
print(df['gender'].unique())

def convert_gender(status):
    if status == 'Male':
        return 0
    elif status == 'Female':
        return 1
    elif status == 'Other':
        return 2
    
df['gender'] = df['gender'].apply(convert_gender)  
print(df['gender'].unique())
```

```python
print(df['ever_married'].unique())

def convert_ever_married(status):
    if status == 'No':
        return 0
    elif status == 'Yes':
        return 1
    
df['ever_married'] = df['ever_married'].apply(convert_ever_married)  
print(df['ever_married'].unique())
```

- Tratamento de outliers
***

```python
cols = ['bmi','avg_glucose_level']
Q1 = df[cols].quantile(0.25) 
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

condition = ~((df[cols] < (Q1 - 1 * IQR)) | (df[cols] > (Q3 + 1 * IQR))).any(axis=1)

df = df[condition]
```

***
- Grid Search com support vector machine
***

```python
x = df.drop(['stroke'], axis = 1)
y = df['stroke'].to_frame()
   
param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[0.1,0.01,0.001],'kernel':['rbf']}
x_train, x_test, y_train, y_test = train_test_split(x, np.ravel(y), test_size = 0.5, random_state = 2021)

grid = GridSearchCV(SVC(random_state = 2021), param_grid, refit = True, verbose = 3)
grid.fit(x_train,y_train)

print(grid.best_params_)
print(grid.best_estimator_)

predictions = grid.predict(x_test)
print(classification_report(y_test, predictions))

print("%0.2f accuracy" %(accuracy_score(y_test, predictions)))
```

- Grid Search com decision tree
***

```python
x = df.drop(['stroke'], axis = 1)
y = df['stroke'].to_frame()
   
x_train, x_test, y_train, y_test = train_test_split(x, np.ravel(y), test_size = 0.5, random_state = 2021)

tree_param = {'criterion':['gini','entropy'],'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50,70,90,120,150]}

grid = GridSearchCV(DecisionTreeClassifier(), tree_param, cv=5)
grid.fit(x_train,y_train)

print(grid.best_params_)
print(grid.best_estimator_)

predictions = grid.predict(x_test)
print(classification_report(y_test, predictions))

print("%0.2f accuracy" %(accuracy_score(y_test, predictions)))
```
