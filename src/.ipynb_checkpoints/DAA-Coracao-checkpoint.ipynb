# Tratamento de dados

***
- Remo√ß√£o dos duplicados
***

```python
df = df.drop_duplicates()
```

***
- Tratamento de outliers
***

```python
cols = ['trestbps', 'chol', 'fbs', 'thalach', 'oldpeak', 'ca', 'thal']
Q1 = df[cols].quantile(0.25) 
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

condition = ~((df[cols] < (Q1 - 1 * IQR)) | (df[cols] > (Q3 + 1 * IQR))).any(axis=1)

df = df[condition]
```

***
- Teste qui-quadrado
***

```python

X = df.iloc[:,1:10]  #independent columns
y = df.iloc[:,-1]    

bestfeatures = SelectKBest(score_func=chi2, k=9)
fit = bestfeatures.fit(X,y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)


featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']

print(featureScores.nlargest(9,'Score')) 
```

# Treino do modelo

***
- Grid Search com support vector machine
***

```python
x = df.drop(['target'], axis = 1)
y = df['target'].to_frame()
   
param_grid = {'C':[0.1,1,10,100], 'gamma':[0.1,0.01,0.001],'kernel':['rbf']}
x_train, x_test, y_train, y_test = train_test_split(x, np.ravel(y), test_size = 0.5, random_state = 2021)

grid = GridSearchCV(SVC(random_state = 2021), param_grid, refit = True, verbose = 3)
grid.fit(x_train,y_train)

print(grid.best_params_)
print(grid.best_estimator_)

predictions = grid.predict(x_test)
print(classification_report(y_test, predictions))

print("%0.2f accuracy" %(accuracy_score(y_test, predictions)))
```

```python
sns.set_style('whitegrid')
sns.countplot(x='target',data=pd.DataFrame(y_train,columns=['target']),palette='rainbow')
```

```python
plot_confusion_matrix(grid, x_test, y_test)
```

***
- Grid Search com decision tree
***

```python
x = df.drop(['target'], axis = 1)
y = df['target'].to_frame()
   
x_train, x_test, y_train, y_test = train_test_split(x, np.ravel(y), test_size = 0.5, random_state = 2021)

tree_param = {'criterion':['gini','entropy'],'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50,70,90,120,150]}

grid = GridSearchCV(DecisionTreeClassifier(), tree_param, cv=5)
grid.fit(x_train,y_train)

print(grid.best_params_)
print(grid.best_estimator_)

predictions = grid.predict(x_test)
print(classification_report(y_test, predictions))

print("%0.2f accuracy" %(accuracy_score(y_test, predictions)))
```

```python
sns.set_style('whitegrid')
sns.countplot(x='stroke',data=pd.DataFrame(y_train,columns=['stroke']),palette='rainbow')
```

```python
plot_confusion_matrix(grid, x_test, y_test)
```
