# $\textbf{Discretização de variáveis nominais}$

### $\textbf{Sex}$

```python
def convertSex(val):
    if val == 'M':
        return 0
    elif val == 'F':
        return 1
    else:
        return -1

df['Sex'] = df['Sex'].apply(convertSex)
print(df['Sex'].unique())
```

### $\textbf{ChestPainType}$

```python
def convertChestPainType(val):
    if val == 'ATA':
        return 0
    elif val == 'NAP':
        return 1
    elif val == 'ASY':
        return 2
    elif val == 'TA':
        return 3
    else:
        return -1
        
df['ChestPainType'] = df['ChestPainType'].apply(convertChestPainType)
print(df['ChestPainType'].unique())
```

### $\textbf{RestingECG}$

```python
def convertRestingECG(val):
    if val == 'Normal':
        return 0
    elif val == 'ST':
        return 1
    elif val == 'LVH':
        return 2
    else:
        return -1

df['RestingECG'] = df['RestingECG'].apply(convertRestingECG)
print(df['RestingECG'].unique())
```

### $\textbf{RestingExerciseAngina}$

```python
def convertRestingExerciseAngina(val):
    if val == 'N':
        return 0
    elif val == 'Y':
        return 1
    else:
        return -1

df['ExerciseAngina'] = df['ExerciseAngina'].apply(convertRestingExerciseAngina)
print(df['ExerciseAngina'].unique())
```

### $\textbf{ST Slope}$

```python
def convertST_Slope(val):
    if val == 'Up':
        return 0
    elif val == 'Flat':
        return 1
    elif val == 'Down':
        return 2
    else:
        return -1

df['ST_Slope'] = df['ST_Slope'].apply(convertST_Slope)
print(df['ST_Slope'].unique())
```

### $\textbf{Teste do Qui-quadrado}$

```python
cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak', 'ST_Slope', 'RestingECG', 'FastingBS', 'ExerciseAngina', 'ChestPainType', 'Sex']

#Hipotese nula (H0): A coluna HeartDiseases e a coluna em questao são indepedentes

def teste_chi2(tabela, intervalo_confianca):
    _, p, _, expected = chi2_contingency(tabela)
    if p <= (1-intervalo_confianca):
        print("Sao dependentes (rejeita-se hipotese nula)")
    else:
        print("Sao independentes (não se rejeita hipotese nula)")
    return expected



for col in cols:
    print("\nTeste qui-quadrado para a coluna: " + col)
        
    valor = teste_chi2(pd.crosstab(df['HeartDisease'], df[col], margins=True), 0.95)
```

```python
def hipertensao(val):
    if val > 140:
        return 1
    else:
        return 0

df['HYPERTENSION'] = df['RestingBP'].apply(hipertensao)

df.describe()
```

# $\textbf{Rede neuronal}$

```python
x = df.drop(['HeartDisease'], axis = 1)
y = df['HeartDisease'].to_frame()

param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[0.1,0.01,0.001],'kernel':['rbf']}
```

```python

sns.set_style('whitegrid')
sns.countplot(x='HeartDisease',data=pd.DataFrame(y_train, columns=['HeartDisease']),palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='HeartDisease',data=pd.DataFrame(y_test, columns=['HeartDisease']),palette='rainbow')
```

```python
def build_model(activation = 'relu', learning_rate = 0.01):
    model = Sequential()
    model.add(Dense(16, input_dim = 8, activation = activation))
    model.add(Dense(8, activation = activation))
    model.add(Dense(1, activation = 'sigmoid'))

    model.compile(
        loss = 'binary_crossentropy',
        optimizer = tf.optimizers.Adam(learning_rate), 
        metrics = ['accuracy']
    )
    
    return model

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2021)

TUNING_DICT = {
    'activation': ['relu', 'sigmoid'],
    'learning_rate': [0.01, 0.001]
}

kf = KFold(n_splits = 10, shuffle = True, random_state = 2021)

model = KerasClassifier(build_fn = build_model, epochs = 20, batch_size = 20, verbose=0)

grid_search = GridSearchCV(estimator = model,
                           param_grid = TUNING_DICT,
                           cv = kf,
                           refit = 'True',
                           verbose = 0)

grid_search.fit(x_train, y_train, validation_split = 0.2)
```

```python
best_model = grid_search.best_estimator_

predictions = best_model.predict(x_test)

confusion_matrix(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```
