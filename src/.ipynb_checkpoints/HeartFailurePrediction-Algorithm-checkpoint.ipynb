# $\textbf{Conversão de Métricas}$

### $\textbf{Sex}$

```python
def convertSex(val):
    if val == 'M':
        return 0
    elif val == 'F':
        return 1
    else:
        return -1

df['Sex'] = df['Sex'].apply(convertSex)
print(df['Sex'].unique())
```

### $\textbf{ChestPainType}$

```python
def convertChestPainType(val):
    if val == 'ATA':
        return 0
    elif val == 'NAP':
        return 1
    elif val == 'ASY':
        return 2
    elif val == 'TA':
        return 3
    else:
        return -1
        
df['ChestPainType'] = df['ChestPainType'].apply(convertChestPainType)
print(df['ChestPainType'].unique())
```

### $\textbf{RestingECG}$

```python
def convertRestingECG(val):
    if val == 'Normal':
        return 0
    elif val == 'ST':
        return 1
    elif val == 'LVH':
        return 2
    else:
        return -1

df['RestingECG'] = df['RestingECG'].apply(convertRestingECG)
print(df['RestingECG'].unique())
```

### $\textbf{RestingExerciseAngina}$

```python
def convertRestingExerciseAngina(val):
    if val == 'N':
        return 0
    elif val == 'Y':
        return 1
    else:
        return -1

df['ExerciseAngina'] = df['ExerciseAngina'].apply(convertRestingExerciseAngina)
print(df['ExerciseAngina'].unique())
```

### $\textbf{ST\_Slope}$

```python
def convertST_Slope(val):
    if val == 'Up':
        return 0
    elif val == 'Flat':
        return 1
    elif val == 'Down':
        return 2
    else:
        return -1

df['ST_Slope'] = df['ST_Slope'].apply(convertST_Slope)
print(df['ST_Slope'].unique())
```

# $\textbf{Treino do Modelo}$

```python
def model(df):
    
    x = df.drop(['HeartDisease'], axis = 1)
    y = df['HeartDisease'].to_frame()
    
    param_grid = {'C':[0.1,1,10,100], 'gamma':[0.1,0.01,0.001],'kernel':['rbf']}
    x_train, x_test, y_train, y_test = train_test_split(x, np.ravel(y), test_size = 0.3, random_state = 2021)

    grid = GridSearchCV(SVC(random_state = 2021), param_grid, refit = True, verbose = 0)
    grid.fit(x_train,y_train)

    print(grid.best_params_)
    print(grid.best_estimator_)

    predictions = grid.predict(x_test)
    # print(classification_report(y_test, predictions))

    # print("%0.2f accuracy" %(accuracy_score(y_test, predictions)))

    return [grid, x_test, y_test, accuracy_score(y_test, predictions)]
```

```python
import itertools

cols = ['Age', 'Sex', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG', 'MaxHR', 'Oldpeak', 'CholesterolHigh']

all_combinations = []

for r in range(0, len(cols) + 1):
    combinations_object = itertools.combinations(cols, r)
    combinations_list = list(combinations_object)
    all_combinations += combinations_list

print(len(all_combinations))
```

```python
import threading

accuracy = 0
bestPar = ()
bestCols = []

lock = threading.RLock()

def superGrid(combs, start, end):
    global bestPar, bestCols, accuracy

    for elem in combs[start:end]:
        df_test = df_drop.copy()
        df_test = df_test.drop(np.asarray(elem), axis = 1)
        grid, x_test, y_test, acc = model(df_test)
        
        with lock:
            if acc > accuracy:
                bestPar = (grid, x_test, y_test, acc)
                bestCols = np.asarray(elem)
                accuracy = acc
                  

def split_processing(items, num_splits = 20):                                      
    split_size = len(items) // num_splits                                       
    threads = []

    for i in range(num_splits):                                                        
        start = i * split_size                                                  

        end = None if i + 1 == num_splits else (i + 1) * split_size                 
                            
        threads.append(threading.Thread(target = superGrid, args = (items, start, end)))       
        threads[-1].start()
                                          
    for t in threads:                                                           
        t.join()

    print("Accuracy: %0.2f" % accuracy)
    print("Best Columns: ", bestCols)
    plot_confusion_matrix(bestPar[0], bestPar[1], bestPar[2])
```

```python
split_processing(all_combinations)
```
