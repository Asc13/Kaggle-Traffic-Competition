## $\textbf{Discretização de variáveis nominais}$

```python
df_training_convert = df_training_features.copy()
```

### $\textbf{Speed}$

```python
def convertSpeed(speed):
    if speed == 'None':
        return 0
    elif speed == 'Low':
        return 1
    elif speed == 'Medium':
        return 2
    elif speed == 'High':
        return 3
    elif speed == 'Very_High':
        return 4

df_training_convert['AVERAGE_SPEED_DIFF'] = df_training_convert['AVERAGE_SPEED_DIFF'].apply(convertSpeed)
df_training_convert.head()
```

### $\textbf{Cloudiness}$

```python
def convertCloudiness(cloud):
    if cloud == 'c�u pouco nublado' or cloud == 'algumas nuvens':
        return 1
    elif cloud == 'nuvens dispersas' or cloud == 'nuvens quebrados' or cloud == 'nuvens quebradas':
        return 2
    elif cloud == 'nublado' or cloud == 'tempo nublado':
        return 3
    else:
        return 0
    
df_training_convert['AVERAGE_CLOUDINESS'] = df_training_convert['AVERAGE_CLOUDINESS'].apply(convertCloudiness)
```

### $\textbf{Luminosity}$

```python
def convertLuminosity(lum):
    if lum == 'DARK':
        return 0
    elif lum == 'LOW_LIGHT':
        return 1
    elif lum == 'LIGHT':
        return 2

df_training_convert['LUMINOSITY'] = df_training_convert['LUMINOSITY'].apply(convertLuminosity)    
```

### $\textbf{AverageRain}$

```python
def convertAverageRain(rain):
    if rain == 'chuva fraca' or rain == 'chuvisco fraco' or rain == 'chuvisco e chuva fraca' or rain ==  'aguaceiros fracos' or rain ==  'chuva leve':
        return 1
    elif rain == 'chuva':
        return 2
    elif rain == 'aguaceiros':
        return 3
    elif rain == 'chuva moderada':
        return 4
    elif rain == 'chuva forte' or rain ==  'chuva de intensidade pesada' or rain == 'chuva de intensidade pesado':
        return 5
    elif rain == 'trovoada com chuva leve':
        return 6
    elif rain == 'trovoada com chuva':
        return 7
    else:
        return 0
    
df_training_convert['AVERAGE_RAIN'] = df_training_convert['AVERAGE_RAIN'].apply(convertAverageRain)    
```

### $\textbf{Teste do Qui-quadrado}$

```python
cols = ['AVERAGE_TIME_DIFF', 'HOUR', 'LUMINOSITY', 'AVERAGE_HUMIDITY', 'AVERAGE_TEMPERATURE', 'AVERAGE_WIND_SPEED', 'AVERAGE_FREE_FLOW_SPEED', 
        'WEEKEND', 'AVERAGE_FREE_FLOW_TIME', 'DAY', 'AVERAGE_RAIN', 'AVERAGE_CLOUDINESS', 'WEEK', 'MONTH', 'AVERAGE_ATMOSP_PRESSURE']

#Hipotese nula (H0): A coluna HeartDiseases e a coluna em questao são indepedentes

def teste_chi2(tabela, intervalo_confianca):
    stat, p, dof, expected = chi2_contingency(tabela)
    if p <= (1-intervalo_confianca):
        print("Sao dependentes (rejeita-se hipotese nula)")
    else:
        print("Sao independentes (não se rejeita hipotese nula)")
    return expected



for col in cols:
    print("\nTeste qui-quadrado para a coluna: " + col)
        
    valor = teste_chi2(pd.crosstab(df_training_cleaned['AVERAGE_SPEED_DIFF'], df_training_cleaned[col], margins=True), 0.95)
```

```python

#dummy_y = pd.get_dummies(y)
dummy_y = tensorflow.keras.utils.to_categorical(y)

def build_model(activation = 'relu', learning_rate = 0.01):
    model = Sequential()
    model.add(Dense(8, input_dim = 9, activation = activation))
    #model.add(Dense(8, activation = activation))
    model.add(Dense(5, activation = 'softmax'))

    model.compile(
        loss = 'categorical_crossentropy',
        optimizer = tf.optimizers.Adam(learning_rate), 
        metrics = ['accuracy']
    )
    
    return model

dummy_y

```
